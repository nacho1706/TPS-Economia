```python
# Importamos las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<h3>Proponer una función de demanda lineal (Q = a + bP)</h3>

```python
# Proponemos Q = 100 - 2P, donde Q es la cantidad demandada y P es el precio
def demanda(P, a=100, b=-2):
    """
    Función de demanda lineal: Q = a + bP
    Parámetros:
        P: Precio
        a: Intercepto (máxima cantidad cuando P=0)
        b: Pendiente (efecto del precio en la cantidad)
    Retorna:
        Q: Cantidad demandada
    """
    Q = a + b * P
    return Q

# Verificar la función con un ejemplo
P_ejemplo = 20
Q_ejemplo = demanda(P_ejemplo)
print(f" =============== EJEMPLO ===============")
print(f"")
print(f"Para un precio P={P_ejemplo}, la cantidad demandada es Q={Q_ejemplo}")
```

<h3>Construccion de la tabla de demanda para D1</h3>

```python
# Generamos valores de P desde 0 hasta 50 (máximo precio cuando Q=0)
P = np.arange(0, 51, 5)  # Precios de 0 a 50, pasos de 5
Q = demanda(P)
tabla_d1 = pd.DataFrame({'Precio (P)': P, 'Cantidad (Q)': Q})
print("Tabla de Demanda D1:")
display(tabla_d1)
```

<h3>Graficacion de la curva de demanda D1 </h3>

```python
 
plt.figure(figsize=(8, 6))
plt.plot(tabla_d1['Cantidad (Q)'], tabla_d1['Precio (P)'], label='D1', color='blue')
plt.title('Curva de Demanda D1')
plt.xlim(0, 200)
plt.ylim(0, 100)
plt.xlabel('Cantidad (Q)')
plt.ylabel('Precio (A)')
plt.grid(True)
plt.legend()
plt.show()
```

<h3>Graficacion de movimiento a lo largo de la curva D1 (de punto A a punto B)</h3>

```python
# Seleccionamos dos puntos sobre D1
punto_A = {'P': 40, 'Q': demanda(40)}  # Precio alto
punto_B = {'P': 20, 'Q': demanda(20)}  # Precio bajo

# Graficar D1 con los puntos A y B
plt.figure(figsize=(8, 6))
plt.plot(tabla_d1['Cantidad (Q)'], tabla_d1['Precio (P)'], label='D1', color='blue')
plt.scatter([punto_A['Q'], punto_B['Q']], [punto_A['P'], punto_B['P']], color='red', s=100)
plt.text(punto_A['Q'], punto_A['P']+2, 'A', fontsize=12, color='red')
plt.text(punto_B['Q'], punto_B['P']+2, 'B', fontsize=12, color='red')
plt.title('Movimiento a lo largo de la curva D1 (A a B)')
plt.xlim(0, 200)
plt.ylim(0, 100)
plt.xlabel('Cantidad (Q)')
plt.ylabel('Precio (P)')
plt.grid(True)
plt.legend()
plt.show()

print(f"Punto A: Precio={punto_A['P']}, Cantidad={punto_A['Q']}")
print(f"Punto B: Precio={punto_B['P']}, Cantidad={punto_B['Q']}")
```

<h3>Desplazamiento hacia arriba y derecha (D1 a D2)</h3>

```python
# D2: Aumentamos el intercepto (a) para simular un aumento en la demanda
# Nueva función: Q = 120 - 2P (aumentamos a de 100 a 120)
def demanda_d2(P):
    return demanda(P, a=120, b=-2)

# Crear tabla para D2
Q_d2 = demanda_d2(P)
tabla_d2 = pd.DataFrame({'Precio (P)': P, 'Cantidad (Q)': Q_d2})

# Mostrar la tabla
print("Tabla de Demanda D2:")
display(tabla_d2)

# Graficar D1 y D2
plt.figure(figsize=(8, 6))
plt.plot(tabla_d1['Cantidad (Q)'], tabla_d1['Precio (P)'], label='D1', color='blue')
plt.plot(tabla_d2['Cantidad (Q)'], tabla_d2['Precio (P)'], label='D2', color='green')
plt.title('Desplazamiento de la curva de demanda (D1 a D2)')
plt.xlim(0, 200)
plt.ylim(0, 100)
plt.xlabel('Cantidad (Q)')
plt.ylabel('Precio (P)')
plt.grid(True)
plt.legend()
plt.show()
```

<h3>Desplazamiento hacia abajo e izquierda (D1 a D3)</h3>

```python
# D3: Disminuimos el intercepto (a) para simular una disminución en la demanda
# Nueva función: Q = 80 - 2P (disminuimos a de 100 a 80)
def demanda_d3(P):
    return demanda(P, a=80, b=-2)

# Crear tabla para D3
Q_d3 = demanda_d3(P)
tabla_d3 = pd.DataFrame({'Precio (P)': P, 'Cantidad (Q)': Q_d3})

# Mostrar la tabla
print("Tabla de Demanda D3:")
display(tabla_d3)

# Graficar D1 y D3 con una escala ajustada
plt.figure(figsize=(8, 6))
plt.plot(tabla_d1['Cantidad (Q)'], tabla_d1['Precio (P)'], label='D1', color='blue')
plt.plot(tabla_d3['Cantidad (Q)'], tabla_d3['Precio (P)'], label='D3', color='orange')
plt.title('Desplazamiento de la curva de demanda (D1 a D3)')
plt.xlim(0, 200)
plt.ylim(0, 100)
plt.xlabel('Cantidad (Q)')
plt.ylabel('Precio (P)')
plt.grid(True)
plt.legend()
plt.show()
```

<h2>DESAFIO 1</h2>

```python
from ipywidgets import interact, FloatSlider

# Función interactiva para graficar la curva de demanda con un intercepto ajustable
def graficar_demanda_interactiva(a):
    Q_interactiva = demanda(P, a=a, b=-2)
    plt.figure(figsize=(8, 6))
    plt.plot(Q_interactiva, P, label=f'Demanda (a={a:.2f})', color='purple')
    plt.title('Impacto del cambio en la ordenada al origen (a)')
    plt.xlim(0, 200)
    plt.ylim(0, 100)
    plt.xlabel('Cantidad (Q)')
    plt.ylabel('Precio (P)')
    plt.grid(True)
    plt.legend()
    plt.show()

interact(graficar_demanda_interactiva, a=FloatSlider(value=100, min=50, max=150, step=0.1, description='Ordenada (a)'))
```

<h2>DESAFIO 2 </h2>

```python
def graficar_demanda_pendiente(b):
    Q_interactiva = demanda(P, a=100, b=b)
    plt.figure(figsize=(8, 6))
    plt.plot(Q_interactiva, P, label=f'Demanda (b={b:.2f})', color='green')
    plt.title('Impacto del cambio en la pendiente (b)')
    plt.xlim(0, 200)
    plt.ylim(0, 100)
    plt.xlabel('Cantidad (Q)')
    plt.ylabel('Precio (P)')
    plt.grid(True)
    plt.legend()
    plt.show()

interact(graficar_demanda_pendiente, b=FloatSlider(value=-2, min=-10, max=0, step=1, description='Pendiente (b)'))
```
